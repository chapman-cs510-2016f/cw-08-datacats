#### Our code:
The complex number is a number that can be expressed in the form 
$z=a+ib$

where $a$ and $b$ are real numers, and $i$ is an imaginary number where 
$i^2=-1$

These numbers from the complex set 
$\mathbb{C}$


Taking the points in the complex plane, we can transform the points with the following equation

$z = z^2 + c$

where c is a complex parameter.
Depending on our value of c, the set transforms in very interesting ways, forming patters that some have described as 'chatoic'
This transformed set is known as the Julia set, and in this module, we will explore the Julia set and some of its various transformations.

### Implementation
Our module creates a class called JuliaPlane which uses the method show() to plot a image of the complex plane after its is  transfromedfrom the function Julia. 
The points within the complex plane falls in between the x and y values -2 to 2, with 1000 points inbetween. This will create a grid of fine points to effectively view our plots. The default, and first plot, will be using the complex parameter $-1.037 + 0.17j$ to transform the plane to the Julia set. 

We used the 'bone' colormap which has a blue to white tint. We had played around with the different colormaps and noticed that 'hot' also which really showcases the results nice and distinctly. We noticed that any miscellaneous maps, such as 'jet' or 'prism' makes it difficult to see the results properly. 

The bicubic interpolation helps in smoothing out the plot as opposed to have none. Cubic spline interpolation is used widely in math modeling to create smooth lines between points, so we decided to use it here with our graphs.
